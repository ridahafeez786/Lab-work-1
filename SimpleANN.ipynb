```python
import torch
import torch.nn as nn
```

```python
class my_model(nn.Module):
    def __init__(self,data):
        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(data,3),
            nn.ReLU(),
            nn.Linear(3,1),
            nn.Sigmoid()
        )

    def forward(self,data):
      out=self.network(data)
      return out
```

```python
samples= torch.rand(10,5)
samples.shape
```

```python
obj1=my_model(samples.shape[1])
obj1
```

```python
from sklearn.datasets import make_classification
```

```python
x , y= make_classification(
    n_samples=100, n_features=10, n_classes=2, random_state=42
)
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from torch.utils.data import Dataset, DataLoader
```

```python
class my_dataset_class(Dataset):
  def __init__(self, features,labels):
    self.features=torch.tensor(features,dtype=torch.float32)
    self.labels=torch.tensor(labels,dtype=torch.long)
  def __len__(self):
    return self.features.shape[0]
  def __getitem__(self,index):
    return self.features[index], self.labels[index]
```

```python
train_dataset_01 = my_dataset_class(x_train,y_train)
test_dataset_01 = my_dataset_class(x_test,y_test)
```

```python
train_dataloader_01 = DataLoader(train_dataset_01, batch_size=4, shuffle=True)
test_dataloader_01 = DataLoader(test_dataset_01, batch_size=4, shuffle=True)
```

```python
for x,y in test_dataloader_01:
  print(x)
  print(y)
  print("=")
```

```python
class my_model_1(nn.Module):
    def __init__(self,data):
        super().__init__()

        self.network = nn.Sequential(
            nn.Linear(data, 16),   # 10 input features → 16 nodes
            nn.ReLU(),

            nn.Linear(16, 8),    # 16 nodes → 8 nodes
            nn.ReLU(),

            nn.Linear(8, 4),     # 8 nodes → 4 nodes
            nn.ReLU(),

            nn.Linear(4, 1),     # 4 nodes → 1 output
            nn.Sigmoid()
        )

    def forward(self, data):
        out = self.network(data)
        return out
```

```python
obj_02=my_model_1(x_train.shape[1])
obj_02
```

```python
number_of_epoch=100
learning_rate=0.01
criterion=nn.BCELoss()
optimizer=torch.optim.SGD(obj_02.parameters(),lr=learning_rate)
```

```python
for epoch in range(number_of_epoch):
  total_loss=0

  for data, labels in train_dataloader_01:
    out=obj_02(data)
    labels=labels.float().unsqueeze(1)
    print()
    loss= criterion(out,labels)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    total_loss+=loss.item()
    avg_loss=total_loss/len(train_dataloader_01)
    print(f"Epoch:{epoch+1}, Loss: {avg_loss}")
```
